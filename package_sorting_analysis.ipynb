```python
def sort(width, height, lenght, mass):
    volumen = width * height * lenght
    #conditions
    heavy = "heavy" if mass >= 20 else "light"
    bulky = "bulky" if volumen >= 1000000 or width >= 150 or height >= 150 or lenght >= 150 else "standard"
    #Logic
    if heavy == "heavy" and bulky == "bulky":
        return "REJECTED"
    elif heavy == "heavy" or bulky == "bulky":
        return "SPECIAL"
    elif heavy == "light" and bulky == "standard":
        return "STANDARD"
```

```python
def test_sort():
    assert sort(10, 10, 10, 5) == "STANDARD", "Test case 1 failed"
    assert sort(200, 200, 200, 25) == "REJECTED", "Test case 2 failed"
    assert sort(150, 150, 150, 15) == "SPECIAL", "Test case 3 failed"
    assert sort(50, 50, 50, 30) == "SPECIAL", "Test case 4 failed"
    assert sort(100, 100, 100, 10) == "SPECIAL", "Test case 5 failed"
    print("All test cases pass")
test_sort()
```

```python
# ARTIFACT - raw data
raw_data = """Width,Height,Length,Mass
50,60,70,10
,None,50,5
200,100,50,25
150,150,150,30
60,80,100,-15
85,95,105,20
100,abc,100,40
20,30,40,0
95,105,115,24
100,100,100
45,55,65,12
-50,60,70,10
110,120,130,22
80,90,100,20,0
60,70,80,18
Width,Height,Length,Mass
140,150,160,35
70,80,90,15
,None,None,None
80,90,100,15
None,None,None,None
50,60,70,10
"100","100","100","40"
200,100,50,25
100,,100,40
30,40,50,5
150,150,150,30
55,65,75,14
60,80,100,15
90,100,110,26
100,100,100,40
85,80,75,14
20,30,40,8
70,65,60,12
90,95,100,20
55,50,45,10
65,75,85,16
,None,40,50,5
95,105,115,24
None,None,None,None
32,32,23ab,43
"""

# Classification function


def classify_package(width, height, length, mass):
    volume = width * height * length
    heavy = "heavy" if mass >= 20 else "light"
    bulky = "bulky" if volume >= 1_000_000 or width >= 150 or height >= 150 or length >= 150 else "standard"

    if heavy == "heavy" and bulky == "bulky":
        return "REJECTED"
    elif heavy == "heavy" or bulky == "bulky":
        return "SPECIAL"
    else:
        return "STANDARD"


# Process CSV
lines = raw_data.strip().split('\n')

# Control variables
results = []
errors = []

# Row-by-row processing
for idx, line in enumerate(lines):
    original_line = line.strip()
    fields = [f.strip().replace('"', '') for f in original_line.split(',')]

    # Skip duplicated headers
    if fields == ['Width', 'Height', 'Length', 'Mass']:
        continue

    # Validate the number of columns
    if len(fields) != 4:
        errors.append({
            'line': idx,
            'data': fields,
            'error': f"Incorrect number of columns: {len(fields)}"
        })
        continue

    # Try to convert to numbers
    try:
        width = float(fields[0]) if fields[0].lower(
        ) != 'none' and fields[0] != '' else None
        height = float(fields[1]) if fields[1].lower(
        ) != 'none' and fields[1] != '' else None
        length = float(fields[2]) if fields[2].lower(
        ) != 'none' and fields[2] != '' else None
        mass = float(fields[3]) if fields[3].lower(
        ) != 'none' and fields[3] != '' else None
    except ValueError:
        errors.append({
            'line': idx,
            'data': fields,
            'error': "Failed conversion: non-numeric type"
        })
        continue

    # Validate null fields
    if None in [width, height, length, mass]:
        errors.append({
            'line': idx,
            'data': fields,
            'error': "Null or empty fields detected"
        })
        continue

    # Validate negative values
    if any(val < 0 for val in [width, height, length]) or mass < 0:
        errors.append({
            'line': idx,
            'data': fields,
            'error': "Negative values are not allowed"
        })
        continue

    # If everything is fine, classify the package
    classification = classify_package(width, height, length, mass)
    results.append({
        'line': idx,
        'data': (width, height, length, mass),
        'classification': classification
    })

# Print results
print("=== Classification Results ===\n")
for r in results:
    print(
        f"Row {r['line']}: {r['data']} => Classified as '{r['classification']}'")

print("\n=== Discarded Packages / Internal Processes ===\n")
for e in errors:
    print(f"Row {e['line']}: {e['data']} => ERROR: {e['error']}")
```
